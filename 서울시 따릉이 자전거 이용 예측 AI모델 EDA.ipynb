```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('train.csv')
train_data
```

```python
train_data.isnull().sum()
```

```python
train_data.fillna(0, inplace = True)
```

```python
colormap = plt.cm.gist_heat
plt.figure(figsize = (15, 12))
sns.heatmap(train_data.corr(), linewidths = 0.1, vmax = 0.5, cmap = colormap, linecolor = 'white', annot = True)
```

```python
sns.pairplot(vars = ['hour', 'hour_bef_temperature', 'hour_bef_precipitation', 'hour_bef_windspeed', 'hour_bef_humidity', 'count'], data = train_data)
plt.show()
```

```python
sns.pairplot(vars = ['hour_bef_visibility', 'hour_bef_ozone', 'hour_bef_pm10', 'hour_bef_pm2.5', 'count'], data = train_data)
plt.show()
```

```python
sns.countplot(train_data['hour_bef_precipitation'])
plt.show()
```

```python
sns.boxplot(x = 'hour', data = train_data,)
plt.show()
```

```python
sns.boxplot(x = 'hour_bef_temperature', data = train_data)
plt.show()
```
